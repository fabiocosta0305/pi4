```python
import pandas as pd
import hvplot.pandas  # habilita df.hvplot()
import panel as pn
import dados
import hvplot
pd.set_option('future.no_silent_downcasting', True)
```

```python
source_data=dados.cria_dados()
source_data['lista_cras']
```

```python
def value(cras):
    return cras

def info_cras(dados,cras):
    # print(dados)
    # print(cras)e
    dados_pessoas=dados['pessoas']
    dados_plotar=dados_pessoas[dados_pessoas['Unidade']==cras]
    # tipos=pd.concat([pd.Series('Ms'),dados_plotar['Tipo']])
    # tipos.insert(loc=0,column='Tipo', value='Ms')
    # print(tipos)
    tipos=dados_plotar['Tipo']
    dados_plotar.index=dados_plotar['Tipo']
    dados_plotar=dados_plotar.loc[:,'Janeiro':'Dezembro']
    # print(dados_plotar)
    return dados_plotar   

def graph_cras(dados,cras):
    dados_plotar=info_cras(dados,cras)
    dados_plotarT = dados_plotar.T.reset_index().rename(columns={"index": "Ms"})
    # dados_plotar.index=dados_plotar['index']
    # dados_plotar=dados_plotar.loc[:,'Total':'Coletivo']
    dados_plotar.rename(columns={'index':'Ms'}, inplace=True)
    print(dados_plotarT)
    table = pn.widgets.DataFrame(dados_plotar, name=f"# {cras}",sizing_mode='stretch_both')
    # Cria grfico de barras
    bar_plot = dados_plotarT.hvplot.bar(
        x="Ms", 
        y=dados_plotar.index,   # usa as mtricas como sries
        stacked=False,
        rot=45,
        title=f"# {cras}",
        shared_axes=False,
        label='Ms',
        legend_cols=1,
        # subplots=True,
        legend='top',              # garante exibio
        #legend_position='top_right' # posio da legenda
    )
    return pn.Column(
        f"# {cras}",
        table,
        bar_plot
    )
    
drop_cras = pn.widgets.Select(
    name="CRAS",
    options=sorted(source_data['lista_cras']),
    value=sorted(source_data['lista_cras'])[0],
    description="Lista Completa dos CRAS disponveis",
)

dados_cras=pn.bind(info_cras, dados=source_data, cras=drop_cras)
graph=pn.bind(graph_cras, dados=source_data, cras=drop_cras)

texto_cras=pn.bind(value,cras=drop_cras)

pn.Column(drop_cras,texto_cras,dados_cras,graph)
```
