```python
import pandas as pd
import hvplot.pandas  # habilita df.hvplot()
import panel as pn

# Ativa extenso Panel + HoloViews
pn.extension('hvplot')

import pandas as pd
import hvplot.pandas
import panel as pn

pn.extension('hvplot')

# Exemplo de DataFrame
data = {
    "Jan": [100, 80, 50],
    "Fev": [120, 60, 70],
    "Mar": [90, 100, 60],
    "Abr": [110, 95, 85]
}
index = ["Receita", "Despesas", "Lucro"]
df = pd.DataFrame(data, index=index)

# Transforma para formato "ms" no eixo X
df_t = df.T.reset_index().rename(columns={"index": "Ms"})

# Tabela de visualizao
table = pn.widgets.DataFrame(df_t, name="DataFrame Transformado", width=400, height=200)

# Grfico de barras com meses no eixo X e mtricas na legenda
bar_plot = df_t.hvplot.bar(
    x="Ms", 
    y=df.index,     # cada linha do df original vira uma srie
    stacked=False,
    title="Mtricas por Ms",
    legend='top',               # garante legenda
    legend_position='top_right' # coloca legenda no canto superior direito
)

# Dashboard final
dashboard = pn.Column(
    "# Dashboard Financeiro",
    table,
    bar_plot
)

dashboard.show()
```
