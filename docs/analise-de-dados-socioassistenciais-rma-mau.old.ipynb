```python
import pandas as pd
import matplotlib.pyplot as plt

# Lendo a planilha sem cabeçalho
df_raw = pd.read_excel("../CONSOLIDADO 2024 JAN-DEZ.xlsx",sheet_name='CRAS GERAL', header=None)

# Janeiro até Dezembro (colunas 39 a 50)
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho",
         "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]

col_inicio = 39  # coluna Janeiro
col_fim = 50     # coluna Dezembro

# Pegar as três linhas da seção 1.1 (Total, Individual, Coletivo)
dados = df_raw.iloc[13:16, col_inicio:col_fim+1]

# Definir os índices corretamente
dados.index = ["Total", "Individual", "Coletivo"]
dados.columns = meses

print(dados)

# --- Plot ---
dados.T.plot(kind="bar", figsize=(12,6))
plt.title("Recepção - Quantidade de Pessoas - Cras Geral (2024)", fontsize=14, fontweight="bold")
plt.ylabel("Quantidade de Pessoas")
plt.xticks(rotation=45)
plt.legend(title="Categoria")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar a aba (exemplo: CRAS Geral)
df_raw = pd.read_excel("../CONSOLIDADO 2024 JAN-DEZ.xlsx", sheet_name="CRAS GERAL", header=None)

# Definir meses (colunas 39 a 50 = Janeiro a Dezembro)
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho",
         "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]

col_inicio = 39  # coluna Janeiro
col_fim = 50     # coluna Dezembro

# ---- DEMANDAS (1.4) ----
# Selecionar as linhas da seção 1.4 (exemplo: 83 até 120 - ajuste conforme sua planilha)
demandas = df_raw.iloc[45:76, col_inicio:col_fim+1]

# Pegar os nomes das demandas na coluna A (coluna 0)
nomes_demandas = df_raw.iloc[45:76, 0].tolist()

# Definir colunas e índices
demandas.columns = meses
demandas.index = nomes_demandas

# Somar o total por demanda ao longo do ano
totais_demandas = demandas.sum(axis=1)

# Selecionar top 10 demandas
top_demandas = totais_demandas.sort_values(ascending=False).head(10)

# Plotar gráfico
plt.figure(figsize=(10, 7))
top_demandas.plot(kind="barh", color="steelblue", edgecolor="black")

plt.title("Principais Demandas - CRAS Geral 2024", fontsize=16, fontweight="bold")
plt.xlabel("Número de Atendimentos")
plt.ylabel("Demanda")
plt.gca().invert_yaxis()
plt.grid(axis="x", alpha=0.3)

plt.tight_layout()
plt.show()
```
